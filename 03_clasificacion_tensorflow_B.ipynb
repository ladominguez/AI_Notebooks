```python
# Cargando datos (gato/no-gato)
training_images, training_labels, test_images, test_labels, classes = load_dataset()
training_images = training_images/255
test_images = test_images/255
```

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.95):
      print("\nReached 95% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()
```

```python
#training_images.shape
training_labels=np.squeeze(training_labels.T)
training_labels.shape
```

```python
test_labels=np.squeeze(test_labels.T)
test_labels.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(64, 64, 3)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')  # adjust output units depending on your task
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(training_images, training_labels, epochs=70, callbacks=[callbacks])
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
